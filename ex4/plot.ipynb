```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
benchmarks = ["read", "update", "mixed"]
BENCH = "Benchmark"
IMP = "Implementation"
NOP = "Number of operations"
THREAD = "Threads"
```

```python
#Load csv, plot chart
data = pd.read_csv("ex4_data.csv")
for bm in benchmarks:
    grouped = data[data[BENCH] == bm].pivot(THREAD, IMP, NOP)
    plt.rcParams['figure.figsize'] = [20, 8]
    plt.rcParams.update({'font.size': 22})

    ax = grouped.plot(kind='bar')
    ax.set_ylabel("Number of operations")
    plt.xticks(rotation=90, ha='right')
    plt.savefig(f"plot_{bm}.png", bbox_inches='tight')
```
